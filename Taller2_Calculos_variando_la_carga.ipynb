```python
## a minima carga
Sn0 = (900 - velocidad_cruce_min) / 900
print(Sn0)
```

```python
Rr_min = (R1*R2*(R1+R2)+Sn0**2*(R1*X2**2+R2*X1**2)) / ((R1+R2)**2+Sn0**2*(X1+X2)**2)
Xr_min = (R1**2 * X2 + R2**2 * X1 + Sn0**2 * X1 * X2 * (X1 + X2)) / ((R1 + R2)**2 + Sn0**2 * (X1 + X2)**2)
```

```python
pp1=(1j*Xm1*((Rr_min/Sn0)+1j*Xr_min))/(1j*Xm1+((Rr_min/Sn0)+1j*Xr_min))
Zeq_min=Rs+1j*Xs+(pp1)
r0, theta0 = abs(Zeq_min), cmath.phase(Zeq_min)
print(Zeq_min)
print(f"Forma polar: {r0}∠{theta0} ")
```

```python
Ie_min = 380/r0
print(Ie_min)
```

```python
Pin_N_min = (3 * 380 * Ie_min) * (np.cos(theta0))
print(Pin_N_min)
```

```python
Pcu_E_min = (3 * (Ie_min**2) * Rs )
print(Pcu_E_min)
```

```python
Vth = 380*(((Xm1)/((Rs**2)+((Xs+Xm1)**2))**0.5))
print(Vth)
```

```python
Zth = (((complex(0, Xm1))*(Rs + complex(0, Xs))/(Rs + (complex(0, Xs)+ complex(0, Xm1)))))
Rth = Zth.real
Xth = Zth.imag
```

```python
Ws = ((2*np.pi)/60)*900
print(Ws)
```

```python
Te_N = (((3*(Vth**2))*(Rr_min/Sn0))/((Ws*((Rth + (Rr_min/Sn0))**2+((Xth + Xr_min)**2)))))
print(Te_N)
```

```python
Is_N = 380/Zeq_min
y, z = abs(Is_N), cmath.phase(Is_N)
print(f"Forma polar: {y}∠{z} ")
```

```python
Pin_N = 380*y*np.cos(z)
print(f"Potencia de entrada: {Pin_N} ")
print(f"Factor de potencia a carga minima: {np.cos(z)}")
```
