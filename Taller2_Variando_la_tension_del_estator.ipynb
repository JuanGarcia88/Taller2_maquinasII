```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import cmath
import math


## MEZCLADORA
TmN = 120.54
Vel = 3565

## Con Ve al 100 %
Speed = np.linspace(0, 901, 3000000)
Sn_plot=(900-Speed)/900

f=60
Rs = 0.5907
Xs = (2*np.pi*f)*(0.0009007)
Xm1 = (2*np.pi*f)*(0.03498)
R1 = 0.3492
X1 = (2*np.pi*f)*(0.004182)
R2 = 0.3503
X2 = (2*np.pi*f)*((0.0009007))
Rr_plot= (R1*R2*(R1+R2)+(Sn_plot**2)*(R1*X2**2+R2*X1**2)) / ((R1+R2)**2+(Sn_plot)**2*(X1+X2)**2)
Xr_plot = (R1**2 * X2 + R2**2 * X1 + (Sn_plot**2) * X1 * X2 * (X1 + X2)) / ((R1 + R2)**2 + (Sn_plot**2) * (X1 + X2)**2)
Ws = 94.24777960769379
Vth0 = 370.10827042487097
Rth0 = 0.5603473952635081
Xth0 = 0.35550205971601345
Te_PLOT2 = (((3*(Vth0**2))*(Rr_plot/Sn_plot))/((Ws*((Rth0 + (Rr_plot/Sn_plot))**2+((Xth0 + Xr_plot)**2)))))
x = np.array([0, 3565])
y = np.array([0, 120.54])
y_plot=0.0338*(Speed)
Te_PLOT = (((3*(Vth0**2))*(Rr_plot/Sn_plot))/((Ws*((Rth0 + (Rr_plot/Sn_plot))**2+((Xth0 + Xr_plot)**2)))))
punto_de_cruce = None
for i in range(len(Speed)):
    if Te_PLOT[i] <= y_plot[i]:
        punto_de_cruce = i
        break
if punto_de_cruce is not None:
    velocidad_cruce = Speed[punto_de_cruce]
    torque_cruce = Te_PLOT2[punto_de_cruce]
    print(f"El punto de cruce ocurre en Speed = {velocidad_cruce} y Te_PLOT = {torque_cruce}")

plt.figure(figsize=(11, 8))
plt.xlim(0, 1000)
plt.plot(Speed, Te_PLOT)
plt.xlabel('Velocidad del Motor (Speed)')
plt.ylabel('Torque electrico (Te)')
plt.title('Torque vs. Velocidad del Motor')
plt.plot(x, y)
plt.scatter(880, 457.1868188533887, marker= 'x', color='green', label = 'Torque nominal maquina')
plt.scatter(velocidad_cruce, torque_cruce, marker= 'x', color='red', label = 'Punto de cruce')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
